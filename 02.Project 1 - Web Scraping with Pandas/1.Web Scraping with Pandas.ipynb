## Read .csv from Multiple URLs with Pandas

- LInk: root + season + league

```python
# link strucure
"https://football-data.co.uk/mmz4281/" + "2526" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://football-data.co.uk/mmz4281/"
season = "2526/"
```

### Multiple leagues

```python
# create list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames =[]
# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + season + league + '.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element
frames[0]
frames[4]
```

### Multiple Seasons

```python
# create list of leagues
leagues = ['E0', 'E2', 'E3']
frames =[]
#Looping through multiple seasons
for season in range(15, 21):
# looping through leagues, read multiple csv and append it into a list
    for league in leagues:
        url = root + str(season) + str(season + 1) + '/' + league + '.csv'
        print(url)
        df = pd.read_csv(url)
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 season * 3 leagues)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[0]
frames[len(frames) - 1]
```

## Organize All the Data in Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}
for league in dict_countries:
    frames = []
    for season in range(15,21):
        url = root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv'
        df = pd.read_csv(url)
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames) # storing all seasons of a league in a single dataframe
    dict_historical_data[league] = df_concat # assigning dataframe to a new dictionary
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['German Bundesliga']
```
